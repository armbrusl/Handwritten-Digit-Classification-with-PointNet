```python
import json
import matplotlib.pyplot as plt
```

```python
# Function to load and combine the histories from multiple JSON files
def load_history(json_file_paths):
    parent_directory = '/home/ymos/Documents/coding/HCwPN_data/History/'
    combined_history = {'loss': [], 'val_loss': [], 'accuracy': [], 'val_accuracy': []}
    
    for json_file_path in json_file_paths:
        path = parent_directory + json_file_path
        with open(path, 'r') as file:
            history = json.load(file)
        
        for key in combined_history.keys():
            combined_history[key].extend(history[key])
    
    return combined_history

# Function to plot the history
def plot_history(history):
    # Plotting loss
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history['loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    # Plotting accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history['accuracy'], label='Training Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
json_file_paths = ['20231028-165255.json']
history = load_history(json_file_paths)
plot_history(history)
```
